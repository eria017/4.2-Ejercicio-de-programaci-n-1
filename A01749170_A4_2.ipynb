```python
# Function to calculate the mean
def calculate_mean(data):
  total = sum(data)
  return total / len(data)
```

```python
# Function to calculate the median
def calculate_median(data):
  sorted_data = sorted(data)
  n = len(data)
  middle = n // 2

  if n % 2 == 0:
    return (sorted_data[middle - 1] + sorted_data[middle]) / 2
  return sorted_data[middle]

```

```python
# Function to calculate the mode
def calculate_mode(data):
  frequency =  Counter(data)
  max_frequency = max(frequency.values())
  mode = [key for key, val in frequency.items() if val == max_frequency]

  if len(mode) == len(data):
    return "No hay moda"
  return mode
```

```python
# Function to calculate the variance
def calculate_variance(data, mean):
  squared_diff = [(x - mean) ** 2 for x in data]
  return sum(squared_diff) / len(data)
```

```python
# Function to calculate the standard deviation
def calculate_std_deviation(variance):
  return variance ** 0.5
```

```python
# Function to process multiple files
def process_files(folder_path):
    # Files to search
    file_names = [f"TC{i}.txt" for i in range(1, 8)]

    for file_name in file_names:
        file_path = os.path.join(folder_path, file_name)
        try:
            with open(file_path, 'r') as file:
                file_data = []
                for line in file:
                    try:
                        num = float(line.strip())
                        file_data.append(num)
                    except ValueError:
                        print(f"Dato inválido ignorado en {file_name}: {line.strip()}")

            if file_data:
                # Calculate statistics for the current file
                start_time = time.time()
                mean = calculate_mean(file_data)
                median = calculate_median(file_data)
                mode = calculate_mode(file_data)
                variance = calculate_variance(file_data, mean)
                std_deviation = calculate_std_deviation(variance)
                end_time = time.time()
                elapsed_time = end_time - start_time

                # Show results
                results = (
                    f"""Resultados de estadísticas descriptivas para {file_name}:
Media: {mean:.2f}
Mediana: {median:.2f}
Moda: {mode}
Varianza: {variance:.2f}
Desviación estándar: {std_deviation:.2f}
Tiempo transcurrido: {elapsed_time:.4f} segundos"""
                )
                print(results)

                # Save results in a specific file for each test case
                results_file = os.path.join(folder_path, f"StatisticsResults_{file_name.replace('.txt', '')}.txt")
                with open(results_file, 'w') as output_file:
                    output_file.write(results)
                print(f"Resultados de {file_name} guardados en: {results_file}")
            else:
                print(f"No se encontraron datos válidos en el archivo {file_name}")

        except FileNotFoundError:
            print(f"Error: {file_name} no se encontró en la carpeta especificada.")
        except Exception as e:
            print(f"Error al procesar el archivo {file_name}: {e}")

```

```python
# Function to convert a number to binary (without predefined functions)
def to_binary(n):
    if n == 0:
        return "0"
    binary = ""
    while n > 0:
        binary = str(n % 2) + binary
        n //= 2
    return binary
```

```python
# Function to convert a number to hexadecimal (without predefined functions)
def to_hexadecimal(n):
    hex_chars = "0123456789ABCDEF"
    if n == 0:
        return "0"
    hexadecimal = ""
    while n > 0:
        remainder = n % 16
        hexadecimal = hex_chars[remainder] + hexadecimal
        n //= 16
    return hexadecimal
```

# Problema 3 - Count Words

```python
import os
import time
from google.colab import drive
```

```python
# Mount Google Drive
def mount_drive():
    drive.mount('/content/drive')
```

```python
# Function to count words and calculate their frequency
def count_words(file_path):
    word_count = {}
    try:
        with open(file_path, 'r') as file:
            for line in file:
                words = line.strip().split()  # Split the line into words
                for word in words:
                    word = word.lower()  # Convert the word to lowercase to avoid duplicates
                    # Increment the word count
                    if word in word_count:
                        word_count[word] += 1
                    else:
                        word_count[word] = 1
    except FileNotFoundError:
        print(f"Error: No se encontró el archivo: {file_path}")
    except Exception as e:
        print(f"Error inesperado al procesar el archivo {file_path}: {e}")

    return word_count
```

```python
# Main function
def main():
    # Set up Google Drive
    mount_drive()

    # Path to the folder that contains the files
    folder_path = '/content/drive/MyDrive/Colab Notebooks/Inteligencia Artificial Aplicada /Pruebas de Software y Aseguramiento de la Calidad/Actividad 4.2/P3'

    # Check if the folder exists.
    if not os.path.isdir(folder_path):
        print(f"Error: La carpeta especificada no existe: {folder_path}")
        return

    # List files in the folder
    print("Archivos encontrados en la carpeta:")
    files = os.listdir(folder_path)
    for file_name in files:
        print(f"  - {file_name}")

    # Filter files that match the format 'TC1', 'TC2', etc
    files = [f for f in files if f.startswith("TC") and f.endswith(".txt")]
    if not files:
        print("No se encontraron archivos con el formato esperado (TC1, TC2, etc.).")
        return

    start_time = time.time()

    # Process each file
    for file_name in files:
        file_path = os.path.join(folder_path, file_name)
        print(f"\nProcesando archivo: {file_name}")

        # Count words in the file.
        word_count = count_words(file_path)

        # Display results on screen for the current file
        print(f"\nFrecuencia de palabras en {file_name}:")
        for word, count in word_count.items():
            print(f"Palabra: '{word}', Frecuencia: {count}")

        # Save results in a specific file for each TC
        output_file_name = f"WordCount_{file_name.replace('.txt', '')}_Results.txt"
        output_file_path = os.path.join(folder_path, output_file_name)

        with open(output_file_path, 'w') as output_file:
            output_file.write(f"Frecuencia de palabras en {file_name}:\n")
            for word, count in word_count.items():
                output_file.write(f"Palabra: '{word}', Frecuencia: {count}\n")
            output_file.write("\n")

        print(f"Resultados de {file_name} guardados en: {output_file_path}")

    end_time = time.time()
    elapsed_time = end_time - start_time

    # Display total elapsed time
    print(f"\nTiempo total transcurrido: {elapsed_time:.4f} segundos")

# Run the program
main()
```
